````python

# Otsu in cv
image = cv2.imread('../image/img_1.jpg')
# Convert to grayscale image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculate threshold using Otsu's method
otsu_threshold_value, otsu_thresh = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Print the threshold value
print(f"Otsu's threshold value: {otsu_threshold_value}")

# Display the image processed with Otsu's threshold
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu's Thresholding")
plt.axis('off')

plt.show()
````

otsu, better to deal with the pic with two peaks

````python
#Otsu method from skiamge 

import matplotlib.pyplot as plt
from skimage import io, color
from skimage.filters import threshold_otsu

# Read the image
image = io.imread('../image/img_1.jpg')  

# Convert to grayscale image
gray_image = color.rgb2gray(image) #from white to black 0- 1

# Calculate threshold using Otsu's method
otsu_threshold_value = threshold_otsu(gray_image)

# Apply the threshold
otsu_thresh = gray_image > otsu_threshold_value

# Print the threshold value
print(f"Otsu's threshold value: {otsu_threshold_value}")

# Display the image processed with Otsu's threshold
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu's Thresholding")
plt.axis('off')

plt.show()
````

````python

#Li method in skimage

import matplotlib.pyplot as plt
from skimage import io, color
from skimage.filters import threshold_li

# Read the image
image = io.imread('../image/img_1.jpg')  

# Convert to grayscale image
gray_image = color.rgb2gray(image)

# Calculate threshold using Li's method
li_threshold_value = threshold_li(gray_image)

# Apply the threshold
li_thresh = gray_image > li_threshold_value

# Print the threshold value
print(f"Li's threshold value: {li_threshold_value}")

# Display the image processed with Li's threshold
plt.imshow(li_thresh, cmap='gray')
plt.title("Li's Thresholding")
plt.axis('off')

plt.show()
````

for one peak or complicated 

````python
# adaptive thresholding from cv

# Read the image in grayscale mode
image = cv2.imread('../image/img_1.jpg', cv2.IMREAD_GRAYSCALE)  

# Ensure the image is of type uint8
image = image.astype('uint8')

# Apply adaptive thresholding
adaptive_thresh_mean = cv2.adaptiveThreshold(
    image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

adaptive_thresh_gaussian = cv2.adaptiveThreshold(
    image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Display the original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original ')
plt.axis('off')

# Display the result of adaptive mean thresholding
plt.subplot(1, 3, 2)
plt.imshow(adaptive_thresh_mean, cmap='gray')
plt.title('Adaptive Mean ')
plt.axis('off')

# Display the result of adaptive Gaussian thresholding
plt.subplot(1, 3, 3)
plt.imshow(adaptive_thresh_gaussian, cmap='gray')
plt.title('Adaptive Gaussian ')
plt.axis('off')

plt.show()
````
