contrast,the degree difference between the darkest and lightest

````python
import cv2
import matplotlib.pyplot as plt
# Read the image

image = cv2.imread('../image/img_2.jpg')

# Convert color format from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Turn off the axis
plt.show()
````

````python
# Convert to grayscale image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Turn off the axis

plt.show()
````

````python
# ordinary histogram equalization

import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('../image/img_2.jpg', cv2.IMREAD_GRAYSCALE)


# Perform histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original image and the equalized image
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
````

no parameters set manually

````python
#CLAHE

# Read the image
image = cv2.imread('../image/img_2.jpg', cv2.IMREAD_GRAYSCALE)

# Create a CLAHE object
clahe = cv2.createCLAHE(clipLimit=3, tileGridSize=(8, 8))

# Apply CLAHE
clahe_image = clahe.apply(image)

# Display the original image and the CLAHE processed image
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('CLAHE Image')
plt.imshow(clahe_image, cmap='gray')
plt.axis('off')

plt.show()
````

two parameters needed to be set manually, Clip Limit, Grid Size

````python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('../image/img_2.jpg', cv2.IMREAD_GRAYSCALE)



# Create CLAHE objects with different clip limits
clahe_2 = cv2.createCLAHE(clipLimit=2, tileGridSize=(8, 8))
clahe_20 = cv2.createCLAHE(clipLimit=20, tileGridSize=(8, 8))

# Apply CLAHE with different clip limits
clahe_image_2 = clahe_2.apply(image)
clahe_image_20 = clahe_20.apply(image)

# Display the CLAHE processed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('CLAHE Image (Clip Limit 2)')
plt.imshow(clahe_image_2, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('CLAHE Image (Clip Limit 20)')
plt.imshow(clahe_image_20, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
````

typically 3 to 4, lager cliplimit, higer contrast  
focus more on the outline of cheek bone and details on the scarf

````python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('../image/img_2.jpg', cv2.IMREAD_GRAYSCALE)

# Create a CLAHE object with a clip limit of 20
clahe_20 = cv2.createCLAHE(clipLimit=20, tileGridSize=(8, 8))

# Apply CLAHE with a clip limit of 20
clahe_image_20 = clahe_20.apply(image)

# Display the original and CLAHE processed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('CLAHE Image (Clip Limit 20)')
plt.imshow(clahe_image_20, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
````

````python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('../image/img_2.jpg', cv2.IMREAD_GRAYSCALE)

# Create CLAHE objects with different tile grid sizes
clahe_10_10 = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(10, 10))
clahe_100_100 = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(100, 100))

# Apply CLAHE with different tile grid sizes
clahe_image_10_10 = clahe_10_10.apply(image)
clahe_image_100_100 = clahe_100_100.apply(image)

# Display the CLAHE processed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('CLAHE Image (Tile Grid Size 10x10)')
plt.imshow(clahe_image_10_10, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('CLAHE Image (Tile Grid Size 100x100)')
plt.imshow(clahe_image_100_100, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
````

small grid size, localized contrast - detail  
larger size, global enhancement  
like the second face is smoother
